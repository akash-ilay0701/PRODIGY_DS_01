```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Prodigy InfoTech/Task 1/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_45183.csv')
```

```python
df.head()
```

```python
df.drop(['SpecialNotes',	'TableName',	'Unnamed: 5'], axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Region', 'IncomeGroup'])  # Drop rows with nulls
df.head()
```

```python
df['Region'].unique()
```

```python
income_group_counts = df['IncomeGroup'].value_counts()
income_group_counts
```

```python
# Create the bar chart
plt.figure(figsize=(6,6))  # Adjust figure size as needed
plt.bar(income_group_counts.index, income_group_counts.values)
plt.xlabel('Income Group')
plt.ylabel('Count')
plt.title('Count of Income Groups')
plt.xticks(rotation=30, ha='right')  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a faint grid
plt.tight_layout()
plt.show()
```

```python
region_counts = df['Region'].value_counts()
region_counts
```

```python
# Create the bar chart
plt.figure(figsize=(6,6))  # Adjust figure size as needed
plt.bar(region_counts.index, region_counts.values)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Count of Region')
plt.xticks(rotation=30, ha='right')  # Rotate x-axis labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a faint grid
plt.tight_layout()
plt.show()
```
